# Cleaning the Data Extracted

## Importing the data

```python
import pandas as pd
import numpy as np

historical_data_world_cups_df = pd.read_csv('D:\Proyectos\Soccer_World_Championship\data\historical_data_world_cups.csv')
df_results = pd.read_csv('D:\Proyectos\Soccer_World_Championship\data\fixture.csv')
historical_data_world_cups_df
```

```python
# Cleaning df_results
df_results['Home'] = df_results['Home'].str.strip()
df_results['Away'] = df_results['Away'].str.strip()
df_results
 
```

```python
# Cleaning historical_data_world_cups_df
historical_data_world_cups_df['Home'] = historical_data_world_cups_df['Home'].str.strip()
historical_data_world_cups_df['Away'] = historical_data_world_cups_df['Away'].str.strip()
historical_data_world_cups_df['Score'] = historical_data_world_cups_df['Score'].str.strip()
```

```python
#There is a matche wich wasn't done, so we're gonna drop that match
index_to_drop = historical_data_world_cups_df[historical_data_world_cups_df['Home'].str.contains('Sweden') & 
                                              historical_data_world_cups_df['Away'].str.contains('Austria')].index
historical_data_world_cups_df.drop(index=index_to_drop, inplace=True)
```

```python
# We drop all the letters had been in the Score Column
historical_data_world_cups_df['Score']= historical_data_world_cups_df['Score'].str.replace("[^\d–]","",regex=True)
historical_data_world_cups_df['Score']
```

```python
# We split the score data into two columns for a better procesing 
historical_data_world_cups_df[['HomeGoals', 'AwayGoals']] = historical_data_world_cups_df['Score'].str.split('–', expand=True)
historical_data_world_cups_df.drop('Score', axis=1, inplace=True)
historical_data_world_cups_df 
```

```python
historical_data_world_cups_df.rename(columns={'Home':'HomeTeam',
                                              'Away':'AwayTeam',
                                              'year':'Year'},inplace=True)
historical_data_world_cups_df
```

```python
#historical_data_world_cups_df.dtypes
historical_data_world_cups_df= historical_data_world_cups_df.astype({'HomeGoals':int,
                                      'AwayGoals':int,
                                      'Year':int})
historical_data_world_cups_df.dtypes
```

## Exporting data cleaned

```python
historical_data_world_cups_df.to_csv('D:\Proyectos\Soccer_World_Championship\data\clean_historical_data_world_cups.csv', index=False)
df_results.to_csv('D:\Proyectos\Soccer_World_Championship\data\clean_fifa_worldcup_fixture.csv', index=False)
```
